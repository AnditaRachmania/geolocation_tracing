```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('data_test.csv', index_col=0, parse_dates=['time_entry', 'time_exit'])
```

```python
df.head()
```

```python
df.info()
```

```python
entry_dt = df['time_entry'].dt
exit_dt = df['time_exit'].dt

# extract time_entry as hour, create new column 'entry_hour'
df['entry_hour'] = entry_dt.hour

# extract time_exit as hour, create new column 'exit_hour'
df['exit_hour'] = exit_dt.hour
```

```python
# df.head()
```

```python
df['duration'] = df['time_exit'] - df['time_entry']
```

```python
df.head()
```

```python
df.info()
```

```python
df['duration_s'] = df['duration']/np.timedelta64(1,'s')
```

```python
# drop columns vmax, vmin, vmean (same process as train_data)
df = df.drop(columns=['vmax', 'vmin', 'vmean', 'hash'])
```

```python
df.info()
```

```python
# df.to_csv('edited_test.csv')
```

```python
df[(pd.notnull(df['x_exit']) == True)]['exit_hour'].value_counts()
```

```python
df[(pd.notnull(df['y_exit']) == True)]['exit_hour'].value_counts()
```

```python
# read preprocessed test data
data = pd.read_csv('edited_test.csv', index_col=0)
data = data.set_index('trajectory_id')
```

```python
data.head()
```

```python
data.info()
```

```python
# slicing data to fill (exit at 15.00 above)

part1 = data[(pd.notnull(data['y_exit']) == True)]
part2 = data[(pd.notnull(data['y_exit']) == False)]
```

```python
part1.info()
```

```python
part1.isnull().sum()
```

```python
part2.info()
```

```python
part2.isnull().sum()
```

```python
# since we're going to predict x_exit and y_exit from part2 dataset, columns x_exit and y_exit are deleted

part2.drop(columns=['x_exit', 'y_exit'], inplace=True)
```

```python
part2.head()
```

```python
part2.info()
```

```python
# to predict x and y exit, features type as object are deleted, since it's already extracted and convert to another data type (time entry and exit to entry and exit hour; 
# duration to duration_s)

xy = part2.drop(columns=['time_entry', 'time_exit', 'duration'])
xy.head()
```

```python
xy.info()
```

## Import Model

```python
import joblib
```

### 1. Predict x_exit

```python
model_x = joblib.load('model_x')
```

```python
pred_x = model_x.predict(xy)
```

```python
xy['x_exit'] = pred_x
```

```python
xy.head()
```

### 2. Predict y_exit

```python
model_y = joblib.load('model_y')
```

```python
y = xy.drop(columns=['x_exit'])
```

```python
pred_y = model_y.predict(y)
```

```python
y['y_exit'] = pred_y
```

```python
y.head()
```

### 3. Merge result x_exit and y_exit

```python
result = xy.merge(y, on=['trajectory_id', 'x_entry', 'y_entry', 'entry_hour', 'exit_hour', 'duration_s'])
result
## this is processed dataset from part2 dataframe
```

```python
part1.head()
```

```python
sliced = part1.drop(columns=['time_entry', 'time_exit', 'duration'])
sliced
```

```python
result = result[['x_entry', 'y_entry', 'x_exit', 'y_exit', 'entry_hour', 'exit_hour', 'duration_s']]
result
```

```python
df = pd.concat([sliced, result], axis=0)
df
```
